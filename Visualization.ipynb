```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

The Data used for the visualization exercise is the NYC Restaurant Inspections Dataset. There is one main variable in the dataset, the restaurant score, as well as information on the restaurant (cuisine, location), and the inspection dates.

```python
#read in the data
inspections=pd.read_csv("/Users/Raluca/Documents/project/WebExtract.txt", delimiter=",", header=0, low_memory=False)
```

```python
#print the variable names
print inspections.SCORE.describe()
```

```python
#create the inspection year
clean_inspdate = [r.split(' ')[0] for r in inspections.INSPDATE]
inspections['INSPDATE']=clean_inspdate
inspections['inspyear']=pd.DatetimeIndex(pd.to_datetime(inspections.INSPDATE, infer_datetime_format=True)).year
inspections['inspmonth']=pd.DatetimeIndex(pd.to_datetime(inspections.INSPDATE, infer_datetime_format=True)).month
inspections['CURRENTGRADE'].replace('P', 'Grade Pending', inplace = True)
inspections['CURRENTGRADE'].replace('Z', 'Grade Pending', inplace = True)
```

```python
#drop those observations before 2012 (when the NYC restaurant inspection program began)
inspections = inspections.drop(inspections.index[inspections.inspyear <2012])
```

```python
plt.figure(figsize=(10,5))
plt.scatter(inspections.inspmonth, inspections.SCORE, lw=0, alpha=.1)
plt.xlabel("Inspection Month")
xticks = np.arange(1,13)
plt.xticks(xticks)
plt.ylabel("Restaurant Score")
plt.title("NYC Restaurant Scores Over Time",fontsize='15')
```

```python
#Line Plot of of inspection score over time
inspections['inspdate']=pd.DatetimeIndex(pd.to_datetime(inspections.INSPDATE, infer_datetime_format=True)).date
plt.plot(inspections.SCORE, inspections.inspdate)
```

```python
#the distribution of inspection scores
inspections['SCORE'].hist().set_title('NYC Restaurant Inspections Scores')
```

```python
#Restaurant inspection scores by restaurant grade
inspections.SCORE.hist(by=inspections.CURRENTGRADE)
```
